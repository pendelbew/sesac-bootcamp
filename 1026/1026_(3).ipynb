```python
df = df.rename({'poptotal' : 'total', 'popasian': 'asian'}, axis = 1)
```

```python
df.rename({}, axis=1, inplace = True)
df.rename(columns = {})
```

```python
df
```

```python
df['pertotalasian'] = df['asian'] / df['total'] * 100
```

```python
df
```

```python
df['pertotalasian'].mean()
```

```python
def asianMean(c): #변수가 밖에 있다! 위험하게 작성한 예시임....
    if c > (df['pertotalasian'].mean()):
        return 'large'
    return 'small'
```

```python
def isSize(percent, mean):
    if percent > mean:
        return 'large'
    return 'small'
```

```python
df['asianMean']= df['pertotalasian'].apply(asianMean)
```

```python
df['asianMean'].value_counts()
```

```python
df
```

```python
df['asianMean2']= np.where(df['pertotalasian'] > (df['pertotalasian'].mean()), 'large', 'small')
```
