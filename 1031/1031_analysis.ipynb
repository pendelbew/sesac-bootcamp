# 상점별 연월별 아이템 판매량

```python
sales.groupby('shop_id')['item_cnt_day'].sum()
```

```python
len(sales['shop_id'].unique())
```

```python
sales.groupby(['shop_id', 'y_m'])['item_cnt_day'].sum().sort_values()
```

```python
shop_month = sales.groupby(['shop_id', 'y_m'])['item_cnt_day'].sum().reset_index()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (120, 10))
sns.pointplot(x='y_m', y='item_cnt_day', hue='shop_id', data=shop_month[shop_month['shop_id'] < 5])
```
