# 이상치 삭제

- item_price, item_cnt_day 가 0과 같거나 0보다 작은 데이터 삭제
- 년, 월, 일 date분리하기
- 수입 (income), 연월(y_m), 월일(m_d) 데이터 만들기

```python
sales.drop(sales[sales['item_price'] <= 0].index, inplace=True)
```

```python
sales[sales['item_price'] <= 0]
```

```python
sales.drop(sales[sales['item_cnt_day'] <= 0].index, inplace=True)
```

```python
sales[sales['item_cnt_day'] <= 0]
```

```python
sales.shape
```

```python
sales['income'] = (sales['item_price'] * sales['item_cnt_day'])
```

```python
sales
```

```python
sales['date2'] = pds.to_datetime(sales['date'])
```

```python
sales.info()
```

```python
sales['year'] = sales['date2'].dt.year
```

```python
sales['year']
```

```python
sales['month'] = sales['date2'].dt.month
```

```python
sales['day'] = sales['date2'].dt.day
```

```python
sales['y_m'] = sales['year'].astype(str) + '.' + sales['month'].astype(str)
```

```python
sales['y_m']
```

```python
sales['m_d'] = sales['month'].astype(str) + '.' + sales['day'].astype(str)
```

```python
sales['date2'] = sales['date2'].sort_values()
```

```python
pds.to_datetime(sales['date'], format ='%d.%m.%Y')
```

```python
sales[['day', 'month', 'year']] = sales['date'].str.split('.', expand = True)
```

```python
sales['y_m'] = sales['year'] + '.' + sales['month']
```

```python
sales['m_d'] = sales['month'] + '.' + sales['day']
```

```python
y_m_map = {}

for value, key in enumerate (sales['y_m'].unique()):
    y_m_map[key] = value + 1
```

```python
sales['y_m_map']= sales['y_m'].map(y_m_map)
```

```python
y_m_mapping = {key: value+1 for value, key in enumerate(sales['y_m'].unique())} #comprehension
```

```python
sales['y_m_map2'] = sales['y_m'].map(y_m_mapping)
```

```python
m_d_sort = sales['m_d'].sort_values().unique()
```

```python
m_d_mapping
```

```python
m_d_mapping = {key : value+1 for value, key in enumerate(m_d_sort)}
```

```python
sales['m_d_map'] = sales['m_d'].map(m_d_mapping)
```

```python
sales
```

```python
sales.drop(columns=['y_m', 'm_d'],inplace = True)
```

```python
sales.rename({'y_m_map' : 'y_m', 'm_d_map':'m_d'}, axis = 1, inplace = True)
```

```python
sales.to_csv('sales_pre.csv', index = False)
```
