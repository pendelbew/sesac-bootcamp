```python
data3 = []

for num in tqdm(df['elevatorNo']):
    url = 'https://apis.data.go.kr/openapi/service/ElevatorInformationService/getElevatorView'
    serviceKey = 'gCLmjgZMbPHHz6PrGczbNvDfSRCuK6tNGdbBdtpZYGexF1SiOZ92Lp2wZRhF23AEKfy3WrSvJ/LOc6chgTaMhw=='
    elevator_no = num

    params = {'serviceKey' : serviceKey, 'elevator_no' : elevator_no}
    req = requests.get(url, params = params, verify = False)
    soup = BeautifulSoup(req.text, 'xml')
    
    
    temp = []
    
    temp.append(num)
    try:
        temp.append(soup.item.select('lastInspctDe')[0].text)
    except:
        temp.append('')
    try: 
        temp.append(soup.item.select('inspctInstt')[0].text)
    except:
        temp.append('')
    try:
        temp.append(soup.item.select('lastResultNm')[0].text)
    except:
        temp.append('')
    
    data3.append(temp)
```
