```python
import requests 
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings('ignore')
```

```python
from tqdm import tqdm
```

```python
data_set = []
col_list = ['create_date', 'location_id', 'location_name', 'md101_sn', 'msg', 'send_platform']

for i in tqdm(range(1, 162)):
    url = 'https://apis.data.go.kr/1741000/DisasterMsg3/getDisasterMsg1List'
    serviceKey = 'SgmrRsA%2Bn%2B%2F81ytbTTkOeya48YDYcqlTX%2FcxX%2Fy7Bh1WSJ9t4cPUUNMqaGvIjO%2Bq28Ix9qe%2BMO0Z7KB%2FIGUBfg%3D%3D'
    pageNo = str(i)
    numOfRows = '1000'
    type_1 = 'xml'

    api_url = url + '?' + 'serviceKey=' + serviceKey + '&pageNo' + pageNo + '&numOfRows' + numOfRows + '&type' + type_1 
    
    req = requests.get(api_url, verify = False)
    soup = BeautifulSoup(req.text, 'xml')
    
    for i in range(len(soup.select('create_date'))):
        temp = [] 
        for data in col_list:
            temp.append(soup.select(data)[i].text)
            
        data_set.append(temp)
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(data_set, columns = col_list)
```
