```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
sample_submission = pd.read_csv('sample_submission.csv')
```

---

### Homeplanet

```python
train['Group'] = train['PassengerId'].str[:4]
```

```python
home_group = train.groupby(['Group', 'HomePlanet']).size().reset_index()
```

```python
home_group.info()
```

```python
train[train['HomePlanet'].isnull()]
```

```python
home_dict = {}
for k,v in zip(home_group['Group'], home_group['HomePlanet']):
    home_dict[k] = v
```

```python
home_dict
```

```python
train['HomePlanet'].fillna(train['Group'].map(home_dict), inplace = True)
```

---

```python
train['familyname'] = train['Name'].str.split(' ').str[-1]
```

```python
name_group = train.groupby(['familyname', 'HomePlanet']).size().reset_index()
```

```python
name_dict = {}

for k, v in zip(name_group['familyname'], name_group['HomePlanet']):
    name_dict[k] = v
```

```python
name_dict
```

```python
train['HomePlanet'].fillna(train['familyname'].map(name_dict), inplace = True)
```

---

```python
train.groupby(['Destination', 'HomePlanet']).size()
```

```python
dest_dict = {'55 Cancri e' : 'Europa', 'PSO J318.5-22' : 'Earth', 'TRAPPIST-1e' : 'Earth'}
```

```python
train['HomePlanet'].fillna(train['Destination'].map(dest_dict), inplace = True)
```

```python
train['HomePlanet'].isnull().sum()
```

### VIP

```python
train['VIP'].value_counts(dropna=False)
```

```python
train['VIP'].fillna(False, inplace = True)
```

### Fee

```python
fee_list = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
```

```python
for col in fee_list:
    train[col].fillna(train[col].median(), inplace =True)
```

```python
train['Fee'] = train['RoomService'] + train['FoodCourt'] + train['ShoppingMall'] + train['Spa'] + train['VRDeck']
```

### Cabin

```python
train['deck'] = train['Cabin'].str.split('/').str[0]
```

```python
train['side'] = train['Cabin'].str.split('/').str[-1]
```

```python
train.groupby(['HomePlanet', 'deck']).size()
```

```python
train['deck'].fillna(train['HomePlanet'].map({'Earth': 'G', 'Europa': 'B', 'Mars': 'F'}), inplace = True)
```

```python
train['deck'].isnull().sum()
```

```python
side_group = train.groupby(['Group', 'side']).size().reset_index()
```

```python
side_group_dic = {}
for k, v in zip(side_group['Group'], side_group['side']):
    side_group_dic[k] = v
```

```python
train['side'].fillna(train['Group'].map(side_group_dic), inplace =True)
```

```python
train['side'].fillna('S',inplace = True)
```

### Destination

```python
dest_group = train.groupby(['Group', 'Destination']).size().reset_index()
```

```python
dest_group = dest_group.sort_values(0, ascending=False)
```

```python
dest_group = dest_group.drop_duplicates() # 그룹에 따라 다른 행선지 향하는 애들 있었나? 
```

```python
dest_dict = {}
for g, d in zip(dest_group['Group'], dest_group['Destination']):
    dest_dict[g] = d
```

```python
train['Destination'].fillna(train['Group'].map(dest_dict), inplace = True)
```

```python
train['Destination'].fillna('TRAPPIST-1e', inplace = True)
```

### Cryosleep

```python
train[train['Fee'] == 0.0]
```

```python
train['CryoSleep'] = np.where(train['Fee'] > 0, False, True)
```

### Age

```python
train['Age'].fillna(train['Age'].median(),inplace = True)
```

### 돌리기 전 정리

```python
del_col = ['PassengerId', 'Cabin', 'Name', 'familyname']
```

```python
train = train.drop(del_col, axis =1)
```

### 인코딩

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
```

```python
obj_col = []

for i, j in zip(train.dtypes.index, train.dtypes.values):
    if j == 'object':
        obj_col.append(i)
```

```python
obj_col
```

```python
for col in obj_col:
    train[col] = label.fit_transform(train[col])
```
